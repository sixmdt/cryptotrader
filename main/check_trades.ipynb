```python
import pandas as pd
import time
import matplotlib.pyplot as plt
```

```python
trades = pd.read_csv('trades_data', ',', header=0, index_col=0)
```

```python
def to_time(i):
    return('|'.join(time.ctime(i).split(' ')[1:]))

trades['date']  = list(map(to_time, trades['date']))
```
