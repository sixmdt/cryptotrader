```python
import requests
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline 
```

```python
#coindesk btc one year 
page_link = 'https://api.coindesk.com/v1/bpi/historical/close.json?start=2017-05-10&end=2018-05-10'


r = requests.get(page_link)
params = dict(
    date ='2017-05-10'
)


resp = requests.get(url=page_link, params=params)
data = resp.json()
price = data['bpi']
```

```python
cost = []
date = []

for i in range(0,365):
    cost.append(list(price.items())[i][1])
    date.append(list(price.items())[i][0])
```

```python
df = pd.DataFrame(date)
df['cost'] = cost
```

```python
df.to_csv('btc.csv', sep='\t')
```

```python
plt.plot(date, cost, 'c-')
plt.show()
```
